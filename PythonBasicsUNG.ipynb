```python
#### This section contains data sets


#This is data set of Circle-ID and Circle name saved in dictionary format
circle = {
    'AP':'Andhra Pradesh & Telangana',
    'AS':'Assam','BR':'Bihar & Jharkhand',
    'DL':'Delhi',
    'GJ':'Gujrat',
    'HP':'Himachal Pradesh',
    'HR':'Haryana',
    'JK':'Jammu and Kashmir',
    'KL':'Kerela and Lakshadweep',
    'KA':'Karnataka',
    'KO':'Kolkata',
    'MH':'Maharashtra and Goa',
    'MP':'Madhya Pradesh and Chhattisgarh',
    'MU':'Mumbai',
    'NE':'North East',
    'OR':'Orissa',
    'PB':'Punjab',
    'RJ':'Rajasthan',
    'TN':'Tamil Nadu',
    'UE':'UP (East)',
    'UW':'UP (West)',
    'WB':'West Bengal',
    'GB':'Ghaziabad and Noida'
}


# This is dictionary data set showing the options 1,2 as keys and Prepaid and Postpaid as recharge 
# types values  respaectively
recharge_types = {
    '1' : 'Prepaid', 
    '2' : 'Postpaid'
}


#This is a dictionary dataset for choosing recharge packs
packs = {
    'F':"Full-Recharge-Pack", 
    'T':"Talktime-Pack", 
    'D':"Data-Only-Pack"
}


#This dictioanry is for the options of full recharge 
full_recharge = {
    '250': 'Unlimited calls(local/national) + 1.5GB data/day + 100 SMS/ day. Val = 28 days',
    '400': 'Unlimited calls(local/national) + 3.0GB data/day + 100 SMS/ day. Val = 28 days',
    '450': 'Unlimited calls(local/national) + 1.5GB data/day + 100 SMS/ day. Val = 56 days',
    '500': 'Unlimited calls(local/national) + 1.5GB data/day + 100 SMS/ day. Val = 70 days',
    '550': 'Unlimited calls(local/national) + 1.5GB data/day + 100 SMS/ day. Val = 77 days'
}


#This dcitionary is for the options of talktime
talktime = {
    '100':'Talktime: Rs. 82  Val = 28 days',
    '50':'Talktime: Rs. 39  Val = 28 days',
    '30':'Talktime: Rs. 22  Val = 28 days',
    '20':'Talktime: Rs. 14  Val = 28 days',
    '10':'Talktime: Rs. 7  Val = 28 days'
}


#This dictionary is for the options of data packs
data_packs = {
    '1200':'240 GB data Val= 240 days',
    '600':'72 GB data Val= 70 days',
    '250':'50 GB data Val= 28 days',
    '100':'12 GB data Val= 28 days',
    '50':'6 GB data Val= 28 days'
}


#This dictionary is used to access other dictionary sub divisions
pack_values = {
    'F':full_recharge,
    'T':talktime, 
    'D':data_packs
}


# This dictionary is being used for deciding the payment types
payment_types = {
    '1':'Credit card', 
    '2':'Debit card', 
    '3':'UPI', 
    '4':'Net Banking', 
    '5':'Cash'
}

# This dictionary is used for printing receipt
billing = {
    'Customer-Name':None,
    'Mobile-Number':None,
    'Circle-ID':None,
    'Circle':None,
    'Recharge-Type':None,
    'Pack-Type':None,
    'Plan-Details':None,
    'Recharge-Amount':None,
    'Payment-Method':None,
    'GST':None,
    'Payable-Amount':None
}



#### This section has all the functions 

# This function is used to get the use input and validate if the name entered is valid
def name_validation():
    
    # The while loop will run as long as the conditional expression evaluates to True. Infinite loop
    while True:
        
        # It takes the input from user
        enter_name = input("Please enter the Customer Name : ")
        
        ## Needed to detect the space too. So used split and join
        # .split() will spilt the string entered into list
        without_space = enter_name.split()
        
        # ''.join() joins the list and convert into string. 
        # Here we are removing space and storing in temproary variable to detect the name entered with space
        temp = ''.join(without_space)
        
        # if-else is used as conditional statements that provide you with the decision making
        # The isalpha() method returns True if all the characters are alphabet letters (a-z).
        if temp.isalpha():
            
            # return the value to the place from where the function was called
            # The title() is String Method which is used to convert the first character in each word to Uppercase 
            # and remaining characters to Lowercase in the string and returns a new string.
            return(enter_name.title())
        
        # Printing the error when if condition is not fulfilled
        else:
            print("\n--------------------------------------------------------------")
            print("Error: Name is not valid! \nPlease re-enter the name.")
            print("--------------------------------------------------------------\n")
            
            # The continue statement instructs a loop to continue to the next iteration.
            continue
            
            
        
        
# This function is used to enter the mobile number and validate if number enter is valid         
def mobile_validation():
    
    # The while loop will run as long as the conditional expression evaluates to True. Infinite loop
    while True:
        # It takes the input from user
        enter_mobile = input("Please entered the mobile number you want to recharge : ")
        
        # if-else is used as conditional statements that provide you with the decision making
        # The isnumeric() method returns True if all the characters are numeric (0-9), otherwise False.
        # The len() returns the length of a list, string, dictionary, or any other iterable data format.
        if enter_mobile.isnumeric() and len(enter_mobile) == 10:
            
            # return the value to the function called
            return(enter_mobile)
        
        # Printing the error when if condition is not fulfilled
        else :
            print("\n*******************************************************************")
            print("* Error: Mobile number is invaid. Refer to table and try again *")
            print("*******************************************************************\n")
            
            # The continue statement instructs a loop to continue to the next iteration.
            continue
            
            
            
# This function help display the desired dictionary data in tabular form
# When the function is called min one value has to pass 
def display_table(element, value1 = "", value2 = ""):
    
    #A for loop is used for iterating over a sequence (that is either a list, a tuple, a dictionary, a set, or a string).
    for i,j in element:
        #if i want to print any value i can pass in function
        print(value1 + i + " -> " + value2 + j)
        
    print("-------------------------------------------------------------")
    
    
    
# This function is for entering the circle id and get circle associated with it
def select_circle():
    while(True):
        # Printing opening heading for reference table.
        print("Below is the reference table to find circle id associated with circle\n")
        
        # Calling function to dispay dictionary as reference table.
        display_table(circle.items(), "Circle-ID: ", "Circle: ")
        
        # Takes the input of the circle id added
        select_circle_id = input("Please enter the valid circle ID form the list given below : ")
        
        # Convert upper() method converts all lowercase letters in a string to uppercase and returns the modified string.
        select_circle_id = select_circle_id.upper()
        
        if(select_circle_id in circle):
            # return the value to the function called
            return(select_circle_id)
        
        # Printing the error when if condition is not fulfilled
        else:
            print("\n*******************************************************************")
            print("* Error: Selected Circle ID is Invalid. Refer to table and try again *")
            print("*******************************************************************\n")
            # The continue statement instructs a loop to continue to the next iteration.
            continue
            
            
            
            
def recharge_type():
    while True:
        # Printing opening heading for reference table.
        print("Below is the reference table to select recharge type\n")
        
        # Here we are paasing the items to display table function to print in tabular form
        display_table(recharge_types.items())
        
        # Takes the input of the recharge type options added
        select_recharge_type = input("Please enter the valid option to select recharge type : ")
        
        # Searching input in the dictioanry recharge type
        if(select_recharge_type in recharge_types):
            return select_recharge_type
        
        # Printing the error when if condition is not fulfilled
        else:
            print("\n*******************************************************************")
            print("* Error: Selected option is Invalid. Refer to table and try again *")
            print("*******************************************************************\n")
            # The continue statement instructs a loop to continue to the next iteration.
            continue


            
# Function is to select the prepaid pack Full recharge, Talktime, or Data Pack
def prepaid_packs():
        while True:
            print("Below are the detailed list of the recharge packs for FullTime, Talktime and Data-Only recharges.",'\n')
            
            # Printing Bold heading and table format of the data packs available
            print('\033[1m' + "Full Recharge" + '\033[0m')
            display_table(pack_values['F'].items(),"Rs ")
            
            # Printing Bold heading and table format of the data packs available
            print('\033[1m' + "Talktime" + '\033[0m')
            
            display_table(pack_values['T'].items(), "Rs ")
            # Printing Bold heading and table format of the data packs available
            print('\033[1m' + "Data Packs" + '\033[0m')
            display_table(pack_values['D'].items(), "Rs ")
            
            print("Select the desired options from the following")
            display_table(packs.items())
            
            # It takes the input from user
            select_pack = input("Please enter the desired option for Prepaid type : ")
            
            # Converting the input the the upper case
            select_pack = select_pack.upper()
            
            # Seaching input in dictionary 'packs'
            if select_pack in packs:
                return (select_pack)
            
            # Printing the error when if condition is not fulfilled
            else:
                print("\n*******************************************************************")
                print("* Error: Selected option is Invalid. Refer to table and try again *")
                print("*******************************************************************\n")
                # The continue statement instructs a loop to continue to the next iteration.
                continue
  

                
# Funtion to select the recharge plan (in case of prepaid) based on the pack selection
# Also enter ony recharge amount (in case of postpaid)
# A value is passed while calling the function
def recharge_amount(check_value):
    while True:
        
        # The condition of the value that we got is checked
        if(check_value == 'F' or check_value == 'T' or check_value == 'D'):
            
            # Display data in table
            display_table(pack_values[check_value].items(), "Rs ")
            
            # str() converts the data entered in string
            recharge_value = str(input("Enter the correct recharge value from the list : Rs "))
            
            # Searching the input entered in dictionary pack_values
            if(recharge_value in pack_values[check_value]):
                return (recharge_value)
            
            # Printing the error when if condition is not fulfilled
            else:
                print("\n*******************************************************************")
                print("* Error: Selected Plan does not exist. Refer to table and try again *")
                print("*******************************************************************\n")
                # The continue statement instructs a loop to continue to the next iteration.
                continue
        # else if - we use it when we want to mone than 2  or 2 conditions       
        elif(check_value == '2'):
            
            recharge_value = input("Enter the correct recharge value or bill value : Rs ")
            
            # .isnuemric is used to check if the string has number 
            # int() is used to convert string numbers to integer
            if(recharge_value.isnumeric() and int(recharge_value) > 0):
                return (recharge_value)
            
            # Printing the error when if condition is not fulfilled
            else:
                print("\n*******************************************************************")
                print("* Error: Selected Plan does not exist. Refer to table and try again *")
                print("*******************************************************************\n")
                continue
                # The continue statement instructs a loop to continue to the next iteration.
            

# Function to select the payment method
def select_payment_type():
    while True:
        print("Select the number option corresponsing to Payment Type from list given below")
        
        # function called for displying table
        display_table(payment_types.items())
        
        select_payment = input('Enter the option that you want to select as payment type : ')
        
        # Searching input in dictioanry payment_types
        if(select_payment in payment_types):
            return(select_payment)
        
        else:
            print("\n*******************************************************************")
            print("* Error: Selected Plan does not exist. Refer to table and try again *")
            print("*******************************************************************\n")
            # The continue statement instructs a loop to continue to the next iteration.
            continue
            
            
# Here while loop is used to re run the program
while True:
    # Used to start or end the program
    mob = input("Select '1' to Recharge or Enter any input to Exit : ")
    
    #if 
    if(mob == '1'):          
        
        # Entered valid name is stored and function is called
        customer_name = name_validation()
        print("\nThe name entered is :", customer_name)
        
        # Mobile number is stored and function is called
        mobile_number = mobile_validation()
        print("\nThe mobile phone number entered is : ", mobile_number)
        
        # Circle id is stored and function is called
        circle_id = select_circle()
        print("\nThe circle id entered is : ", circle_id )

        # Recharge type is strored and function in called
        selected_recharge_type = recharge_type()
        print("\nThe recharge type selected is : ", selected_recharge_type)
        
        # If Recharge type return 1 then it is prepaid 
        if(selected_recharge_type == '1'):
            
            # Prepaid function is stored and called
            selected_prepaid_pack = prepaid_packs()
            print("\nThe selected prepaid pack is : ", selected_prepaid_pack)
            
            # Rechrage amount is stroed and called also used value returned form recharge type
            prepaid_amount = recharge_amount(selected_prepaid_pack)
            
            # stored value of prepaid amount in a varible
            amount = prepaid_amount
            
            # stored pack selected in variable
            sel_pack = packs[selected_prepaid_pack]
            
            #stored pack details in variable
            sel_plan = pack_values[selected_prepaid_pack][prepaid_amount]

            print("\nThe prepaid plan selected is : ", prepaid_amount)
            print("\nThe prepaid plan is : ", pack_values[selected_prepaid_pack][prepaid_amount])
            
            
        # This is for the postpaid    
        elif(selected_recharge_type == '2'):
            
            # postpaid amount is stored and recharge function is called
            postpaid_amount = recharge_amount(selected_recharge_type)
            
            # Overwrite to avoid leaking
            sel_pack = "NA"
            # Overwrite to avoid leaking
            sel_plan = "NA"
            
            # Stored postpaid amount in variable 
            amount = postpaid_amount
            print("\nThe recharge amount selected is : ", postpaid_amount)
            
        #payment method is stored and called
        payment_method = select_payment_type()
        print("\nThe selected payment method is : ", payment_method)

       
        
        
        # intialing the value that needs to be print
        billing['Customer-Name'] = customer_name
        billing['Mobile-Number'] = mobile_number
        billing['Circle-ID'] = circle_id
        billing['Circle'] = circle[circle_id]
        billing['Recharge-Type'] = recharge_types[selected_recharge_type]
        billing['Pack-Type'] =  sel_pack
        billing['Plan-Details'] = sel_plan
        billing['Payment-Method'] = payment_types[payment_method]
        billing['Recharge-Amount'] = str(int(amount))
        billing['GST'] = str(0.18 * int(amount))
        billing['Payable-Amount'] = str(float(amount) + float(billing['GST']))

        
        # Bill printing template 
        # Used for space
        print()
        print()
        print()
        print("-----------------------------------------------------------------------------------------------------")
        print("----------------------------------------------------Receipt------------------------------------------")
        print("-----------------------------------------------------------------------------------------------------")
        print("Customer Name         :  " + billing['Customer-Name'])
        print("Mobile Number         :  " + billing['Mobile-Number'])
        print("Circle ID             :  " + billing['Circle-ID'])
        print("Circle                :  " + billing['Circle'])
        print("Recharge Type         :  " + billing['Recharge-Type'])
        print("Type of Plan          :  " + billing['Pack-Type'])
        print("Details of Plan       :  " + billing['Plan-Details'])
        print("Payment Method        :  " + billing['Payment-Method'])
        print("Recharge Amount       :  " + "₹"+billing['Recharge-Amount'] + ' INR')
        print("GST(18%)              :  " + "₹"+ billing['GST'])
        print("Total Amount Payable  :  " + "₹"+ (billing['Payable-Amount'] + ' INR'))
        print("------------------------------------------------------------------------------------------------------")
        print("------------------------------------------------------------------------------------------------------")
        print("------------------------------------------------------------------------------------------------------")
        # Used for space
        print()
        print()
        print()
        
    # if you want to close type anything and press enter     
    else:
        print("*****  The program has been closed  *****")
        break
        
        

        
    
        
    
    
    


            
            
            
```
